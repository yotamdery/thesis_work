### Reading files and fixing format

```python
pakar = pd.read_csv(PAKAR_FOLDER+'/locations.csv', header=None, names=['loc', 'id', 'time_stmp'], infer_datetime_format=True)
pakar.head()
```

```python
pakar['lat'] = pakar['loc'].apply(lambda x: x.split('"')[3])
pakar['lon'] = pakar['loc'].apply(lambda x: x.split('"')[7])
```

```python
pakar['time'] = pd.to_datetime(pakar['time_stmp'], infer_datetime_format=True)
pakar['time'].head()
```

```python
pakar.head()
```

```python
def id2sys(id_str):
    if '-' in str(id_str):
        return 'AND'
    return 'IOS'
```

```python
pakar['soft_sys'] = pakar['id'].apply(id2sys)
```

```python
pakar = pakar[['id', 'lat', 'lon', 'time', 'soft_sys']]
```

## Regular

```python
STOPS_PATH = "/data/overflow/for_noya/Data/Stops/stops_hourly.csv"
STOP_COLUMNS = ['partition_day', 'hour', 'stop_tazid', 'count_imsi', 'imsi_type']
stops_df = pd.read_csv(STOPS_PATH, header=0, names=STOP_COLUMNS, parse_dates=['partition_day'])
```

```python
stops_df
```

```python
# Get visitors numbers at a given hour & day
def get_visitors_number_specific_time(stops_df, date, hour):
    ans = stops_df.loc[(stops_df.partition_day == date) & 
                       (stops_df.hour == hour) & 
                       (stops_df.imsi_type == "Visitor")]
    ans = ans[['stop_tazid', 'count_imsi']]
    ans.set_index('stop_tazid', drop=True, inplace=True)
    return ans
```

```python
passover = get_visitors_number_specific_time(
    stops_df, 
    pd.Timestamp(year=2020, month=4, day=6),
    19
)
passover.columns = ['200406-19']
passover = passover.sort_index()
passover['200407-02'] = get_visitors_number_specific_time(
    stops_df, 
    pd.Timestamp(year=2020, month=4, day=7),
    2
)

passover['200408-12'] = get_visitors_number_specific_time(
    stops_df, 
    pd.Timestamp(year=2020, month=4, day=8),
    12
)
passover['200408-23'] = get_visitors_number_specific_time(
    stops_df, 
    pd.Timestamp(year=2020, month=4, day=8),
    23
)
```

```python
passover
```

```python
passover_TAZ = TAZs.merge(passover, left_on='ID', right_index=True)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10))


passover_TAZ.plot(column='200407-02', 
                           ax=ax1, 
                           legend=True, 
                           cmap='OrRd',
                           legend_kwds={'label': "# visitors staying The night",
                            'orientation': "horizontal"});
ax1.set_title('ינש םוי')
passover_TAZ.plot(column='200408-12', 
                           ax=ax2,  
                           cmap='OrRd',
                           legend=True, 
                           legend_kwds={'label': "# visitors staying The night",
                            'orientation': "horizontal"});
ax2.set_title('יעיבר םוי')
```

```python
passover_TAZ['change'] = passover_TAZ['200408-12'] - passover_TAZ['200406-19']
passover_TAZ['change'][passover_TAZ['change'] < 0 ] = 0
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
passover_TAZ.plot(column='change', 
                           ax=ax, 
                           legend=True, 
                           cmap='OrRd',
                           legend_kwds={'label': "change in visitors # staying The night"});
ax.set_title('חספ')
```
